```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
dtset = pd.read_csv("./insurance/insurance.csv")
```

```python
sns.pairplot(dtset)
plt.show()
```

```python
dtset.head(5)
```

```python
X = dtset[['age', 'bmi', 'children']]
y = dtset['charges']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)

reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
# Ejemplo de predicci√≥n
print(reg.predict(pd.DataFrame({'age': [30], 'bmi': [25], 'children': [2]})))
print(reg.predict(pd.DataFrame({'age': [20], 'bmi': [25], 'children': [1]})))
```

```python
# Grafico con datos de prueba
sns.regplot(x=X_test['age'], y=y_test)
plt.title('Reg lineal - Test set')
plt.xlabel('Edad')
plt.ylabel('Costo seguro')
plt.show()
```

```python
# Grafico con datos de entrenamiento
sns.regplot(x=X_train['age'], y=y_train)
plt.title('Reg lineal - Train set')
plt.xlabel('Edad')
plt.ylabel('Costo seguro')
plt.show()
```

```python
# Grafico con datos predictivos
# Agregar las predicciones al conjunto de prueba
X_test['charges_pred'] = y_pred

# Visualizar la regresi√≥n lineal en el conjunto de prueba
sns.lmplot(x='age', y='charges_pred', data=X_test, line_kws={'color': 'red'})
plt.title('Reg lineal - Test set, prediction data')
plt.xlabel('Edad')
plt.ylabel('Costo del seguro')
plt.show()
```

```python
print('Precisi√≥n en el conjunto de prueba: {:.2f}'.format(reg.score(X_test[['age', 'bmi', 'children']], y_test)))
```
